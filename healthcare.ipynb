```python
import tensorflow as tf
from tensorflow.keras import layers, models




IMG_SIZE = 48
BATCH = 64
EPOCHS = 25

# Load data directly from folders
train_data = tf.keras.utils.image_dataset_from_directory(
    r'D:\DS\final_exam\archive (10)\train',
    image_size=(IMG_SIZE, IMG_SIZE),
    color_mode='grayscale',
    batch_size=BATCH
)

test_data = tf.keras.utils.image_dataset_from_directory(
    r'D:\DS\final_exam\archive (10)\test',
    image_size=(IMG_SIZE, IMG_SIZE),
    color_mode='grayscale',
    batch_size=BATCH
)

# Normalize pixels 0-255 вҶ’ 0-1
train_data = train_data.map(lambda x, y: (x / 255.0, y))
test_data  = test_data.map(lambda x, y: (x / 255.0, y))


model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(48,48,1)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.4),
    layers.Dense(7, activation='softmax')   # 7 emotion classes
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
model.fit(train_data, epochs=EPOCHS, validation_data=test_data)
model.save('emotion_model.h5')
print("Model saved!")
```

```python
import cv2
import mediapipe as mp
import tensorflow as tf
import numpy as np


mpface   = mp.solutions.face_mesh
mp_draw  = mp.solutions.drawing_utils
mesh     = mpface.FaceMesh()

model    = tf.keras.models.load_model('emotion_model.h5')
LABELS   = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']
ALERT_EM = ['Angry', 'Fear', 'Sad']   # clinical alert emotions

face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

video = cv2.VideoCapture(0)
while True:
    suc, img = video.read()
    img1    = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    
    result = mesh.process(img1)
    if result.multi_face_landmarks:
        for x in result.multi_face_landmarks:
            mp_draw.draw_landmarks(img, x, mpface.FACEMESH_TESSELATION)


    gray  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        face = gray[y:y+h, x:x+w]
        face = cv2.resize(face, (48, 48)) / 255.0
        face = face.reshape(1, 48, 48, 1)

        pred    = model.predict(face, verbose=0)
        emotion = LABELS[np.argmax(pred)]
        conf    = np.max(pred)

        color = (0, 0, 255) if emotion in ALERT_EM else (0, 200, 0)
        cv2.rectangle(img, (x,y), (x+w, y+h), color, 2)
        cv2.putText(img, f'{emotion} {conf:.0%}', (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)

        if emotion in ALERT_EM and conf > 0.6:
            cv2.putText(img, '! ALERT', (10, 40),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0,0,255), 3)

    cv2.imshow('Emotion', img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import numpy as np


mphand  = mp.solutions.hands
mp_draw = mp.solutions.drawing_utils
data    = mphand.Hands()

def finger_up(lm, tip, pip):
    return lm[tip].y < lm[pip].y

def get_gesture(lm):
    index  = finger_up(lm, 8, 6)
    middle = finger_up(lm, 12, 10)
    ring   = finger_up(lm, 16, 14)
    pinky  = finger_up(lm, 20, 18)
    thumb  = lm[4].x < lm[3].x   # right hand

    # STOP: all fingers up
    if index and middle and ring and pinky and thumb:
        return 'STOP'
    # HELP: fist (all down)
    if not index and not middle and not ring and not pinky:
        return 'HELP'
    # THUMBS UP: only thumb
    if thumb and not index and not middle:
        return 'THUMBS UP'
    # OK: thumb + index close
    d = np.linalg.norm([lm[4].x - lm[8].x, lm[4].y - lm[8].y])
    if d < 0.05 and middle and ring and pinky:
        return 'OK'
    return ''

video = cv2.VideoCapture(0)
while True:
    suc, img = video.read()
    img1     = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    result   = data.process(img1)

    if result.multi_hand_landmarks:
        for x in result.multi_hand_landmarks:
            mp_draw.draw_landmarks(img, x, mphand.HAND_CONNECTIONS,
                mp_draw.DrawingSpec(color=(0,0,255), thickness=5, circle_radius=5),
                mp_draw.DrawingSpec(color=(0,0,255), thickness=5))

            gesture = get_gesture(x.landmark)
            if gesture:
                color = (0,0,255) if gesture in ['HELP'] else (0,200,0)
                cv2.putText(img, gesture, (30, 80),
                            cv2.FONT_HERSHEY_SIMPLEX, 2, color, 3)

    cv2.imshow('Gesture', img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
